### Tipos de dados e variveis

Um dos recursos mais eficientes de uma linguagem de programao  a manipulao de variveis. Variveis so espaos reservados na memria para armazenar dados que sero utilizados durante a execuo de um programa. Podem ter valores de diversos tamanhos e tipos, como nmeros inteiros, reais, complexos, caracteres ou at mesmo longas frases.

Em Python, existem 5 tipos de variveis:


* Numricas


    * Inteiros (int)
    * Pontos Flutuantes (float)
    * Complexos (complex)
    * Booleanos (bool)
    

* Categricas


    * Strings (str)

A atribuio de um valor a uma varivel  feito atravs do sinal "=", e por isso no  necessrio declarar uma varivel juntamente ao seu tipo e/ou tamanho como  comum a outras linguagens de programao.

#### Pontos Flutuantes

```python
pi = 3.1415
print(pi,',', type(pi))
```

#### Booleanos

```python
flag1 = True
print(flag1,',', type(flag1))
```

#### Complexos

As variveis do tipo complexo exigem que sejam declaradas a sua parte real e imaginria. A unidade imaginria  denotada por "**j**".

```python
z_1 = 3 + 4j
z_2 = 5j - 8
print(z_1, ',', type(z_1))
print(z_2, ',', type(z_2))
```

 possvel verificar qual o valor da parte real ou imaginria de uma varivel complexa.

```python
print('A parte real :',z_1.real)
print('A parte imaginria :',z_1.imag)
```

#### Strings

```python
texto = 'Este  um caso especial, abordaremos com mais detalhes adiante.'
print(texto,',',type(texto))
```

### Operaes bsicas

Na linguagem de programao Python  comum utilizar artifcios matemticos. As operaes matemticas bsicas e seus respectivos operadores esto listados abaixo.

Operador | Funo
 :--- | :---
a + b | adio
a - b | subtrao
a * b | multiplicao
a / b | diviso
a ** b | exponenciao
       |
a // b | diviso inteira
a % b | resto da diviso inteira
divmod(a, b) | retorna (a // b, x % y)
abs(a) | mdulo de a
round(a[, n]) | arredonda nmero a n casas decimais

Quando uma expresso contm mais de um operador, a ordem da avaliao depende da ordem das operaes. Para operadores matemticos, o Python segue a conveno matemtica. O acrnimo PEMDAS (Parenteses / Exponenciao / Multiplicao e Diviso / Adio e Subtrao) pode ser til para lembrar das regras:

* Os Parnteses tm a precedncia mais alta e podem ser usados para forar a avaliao de uma expresso na ordem que voc quiser. Tambm  possvel usar parnteses para facilitar a leitura de uma expresso, como no caso de (minute * 100) / 60, mesmo se o resultado no for alterado.

* A Exponenciao tem a prxima precedncia mais alta, ento 1 + 2**3  9, no 27.

* A Multiplicao e a Diviso tm precedncia mais alta que a Adio e a Subtrao. Assim, 2 * 3 - 1  5, no 4, e 6 + 4 / 2  8, no 5.

* Os operadores com a mesma precedncia so avaliados da esquerda para a direita (exceto na exponenciao). Assim, na expresso degrees / 2 * , a diviso acontece primeiro e o resultado  multiplicado por . Para dividir por 2, voc pode usar parnteses ou escrever degrees / 2 / .
