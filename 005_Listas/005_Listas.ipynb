#### Loop em listas

Podemos iterar uma lista atravs dos seus indces

```python
for i in range(0,len(l)):
    print(i,l[i])
```

```python
for i in l:
    print(i)
```

```python
for i in enumerate(l):
    print(f'index= {i[0]} e item da lista: {i[1]}')
```

```python
l_num = [0,1,2,3,4,5,6]
l_num
```

Lembremos que listas so mutveis

```python
for i in range(0,len(l_num)):
    l_num[i] = l_num[i]**2
l_num
```
